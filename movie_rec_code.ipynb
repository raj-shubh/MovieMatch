```python
movies_df = movies.merge(credits,on='title')
```

```python
movies = movies_df[['movie_id','title','overview','genres','keywords','cast','crew','tagline']]
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
#converting str to list
import ast

def str_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(str_to_list)
movies['keywords'] = movies['keywords'].apply(str_to_list)
```

```python
movies.head(2)
```

```python
#taking top 5 cast and convert text to list
def str_to_list_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(str_to_list_cast)
```

```python
#fetching only director from crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(2)
```

```python
#converting to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['tagline'] = movies['tagline'].apply(lambda x:x.split())
```

```python
# removing space
'Johnny Depp'
'JohnnyDepp'

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['tagline']
```

```python
movies.iloc[0]['tags']
```

```python
# droping extra columns
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# Converting list to str
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend_movie(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_movie('Krull')
```
